```python
# Write your code to make your dataset analysis ready.
##First we make a copy of the original data set and then perform data wrangling on it
hb_df1 = hb_df.copy()
```

```python
## Drop Duplicate values in Dataset
hb_df1.drop_duplicates(inplace = True)
```

```python
## filling null values with 0 in children column
hb_df1['children']=hb_df1['children'].fillna(0)
## filling null values with others in country column
hb_df1['country']=hb_df1['country'].fillna('others')
## filling null values with 0 in agent column
hb_df1['agent']=hb_df1['agent'].fillna(0)
## filling null values with 0 in company column
hb_df1['company']=hb_df1['company'].fillna(0)
```

```python
no_guests.head(5)
```

```python
##Droping the Rows

hb_df1.drop(no_guests.index,axis = 0,inplace = True)
```

```python
##Making is_canceled column more umderastand to read
hb_df1['is_canceled'] = hb_df1['is_canceled'].replace({0: 'not_canceled', 1: 'canceled'})
hb_df1['is_repeated_guest'] = hb_df1['is_repeated_guest'].replace({0: 'not_repeatted', 1: 'repeatted'})
```

```python
##Before adding new columns total count of columns in hb_df1
hb_df1.shape
```

```python
##Making New columns
##First column of total number of people for every booking
hb_df1['total_no_of_people'] = hb_df1['adults']+hb_df1['children']+hb_df1['babies']
## Second coulmn how many nights they stay menas total stay count (week+weekend)
hb_df1['total_nights_stay'] = hb_df1['stays_in_week_nights']+hb_df1['stays_in_weekend_nights']
## Third column for for adults with no children and babies
hb_df1['guests_with_no_kids'] = hb_df1['adults'] *((hb_df1['children'] == 0) & (hb_df1['babies'] == 0))
## Fourth column for only adults who have children or babies
hb_df1['guests_with_kids'] = hb_df1['total_no_of_people'] - hb_df1['guests_with_no_kids']
```

Chart -3 Comparing bookings made by Guests with kids and Guests with no kids for both Hotels

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure you're using the correct DataFrame name
hb_df = pd.read_csv("/content/drive/MyDrive/Hotel Bookings.csv")

# Group by 'hotel' and calculate the sum for 'guests_with_no_kids' and 'guests_with_kids'
sums_by_hotel = hb_df.groupby('hotel')[['guests_with_no_kids', 'guests_with_kids']].sum().reset_index()

# Melt the DataFrame to long format for better plotting
sums_melted = sums_by_hotel.melt(id_vars='hotel', value_vars=['guests_with_no_kids', 'guests_with_kids'],
                                  var_name='Guest_Type', value_name='Count')

# Create a bar plot using Seaborn
plt.figure(figsize=(12, 6))
sns.barplot(x='hotel', y='Count', hue='Guest_Type', data=sums_melted, palette={'guests_with_no_kids': 'skyblue', 'guests_with_kids': 'orange'})

# Labeling and giving title to the chart
plt.xlabel('Hotel', fontsize=15)
plt.ylabel('Count', fontsize=12)
plt.title('Comparison of Total Bookings Made by Guests Without Kids and Guests With Kids for Both Hotels')
plt.legend(title='Guest Type')

# Show the chart
plt.show()

```
